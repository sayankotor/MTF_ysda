```python
def to_image(tensor):
    tensor -= tensor.min()
    tensor /= tensor.max()
    tensor *= 255
    return tensor.astype(np.uint8)

def show_im(img, title):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.imshow(to_image(img))
    ax.set_axis_off()
    plt.title(title)
```
